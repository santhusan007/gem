# Final Code to Scrape Moneycontrol website

```python
ser=Service(executable_path='C:\\edgedriver_win64\\msedgedriver.exe')
driver=webdriver.Edge(service=ser)
driver.get("https://www.moneycontrol.com/stocks/marketstats/fii_dii_activity/index.php")
for i in range (1,10):
    time.sleep(1)
    # click on the month link
    driver.find_element(By.XPATH,f'//*[@id="pficash"]/div/div/table/tbody/tr[{i}]/td[1]/span[1]/a').click()
    time.sleep(3)
    #capture the html elemnt for scrapping
    html=driver.page_source
    soup = BeautifulSoup(html,'lxml')
    tables = soup.find_all('table')
    
    df= pd.read_html(str(tables[5]))[0]
    df.columns = ['date','FII-Gross Purchase','FII-Gross Sales','FII-Net Purchase / Sales','DII-Gross Purchase','DII-Gross Sales','DII-Net Purchase / Sales']                                
            
    
    #if main_df.empty:
     #       main_df=df
        
    if os.path.isfile("fiibuysell.csv"):        
            df.to_csv("fiibuysell.csv",index=False,header=False,mode='a')            
    else:
             df.to_csv("fiibuysell.csv",index=False)
    #else:
      #     main_df=main_df.append(df,ignore_index=True)
   
    driver.find_element(By.XPATH,'//*[@id="myModal"]/div/div/button').click()
driver.close()    
```
